```python
#################################################################################################
#------------------------------------ Missing data imputation-----------------------------------#
#################################################################################################
import numpy as np
import pandas as pd
import imputation as impt

train = pd.read_csv('../input/train.csv')
test = pd.read_csv('../input/test.csv')
```

```python
# Drop features that have too many missing values
train.drop(['ps_car_03_cat', 'ps_car_05_cat'], axis=1, inplace=True)
test.drop(['ps_car_03_cat', 'ps_car_05_cat'], axis=1, inplace=True)

# One hot encoding for non-missing categorical columns
columns = ['ps_car_04_cat', 'ps_car_06_cat', 'ps_car_10_cat', 'ps_car_11_cat']
train = pd.get_dummies(train, columns=columns)
test = pd.get_dummies(test, columns=columns)

# Impute ps_ind_04_cat
columns_miss = ['ps_ind_02_cat', 'ps_ind_05_cat', 'ps_reg_03', 
                'ps_car_01_cat', 'ps_car_02_cat', 'ps_car_07_cat', 'ps_car_09_cat', 
                'ps_car_11', 'ps_car_14']
impt.randomForestRegression(train, test, columns_miss, 'ps_ind_04_cat', fill_train=False)
impt.randomForestRegression(train, train, columns_miss, 'ps_ind_04_cat')

# Impute ps_car_02_cat
columns_miss = ['ps_ind_02_cat', 'ps_ind_04_cat', 'ps_ind_05_cat', 'ps_reg_03', 
                'ps_car_01_cat', 'ps_car_07_cat', 'ps_car_09_cat', 
                'ps_car_11', 'ps_car_14']
impt.randomForestRegression(train, test, columns_miss, 'ps_car_02_cat', fill_train=False)
impt.randomForestRegression(train, train, columns_miss, 'ps_car_02_cat')

# Impute ps_car_11
columns_miss = ['ps_ind_02_cat', 'ps_ind_04_cat', 'ps_ind_05_cat', 'ps_reg_03', 
                'ps_car_01_cat', 'ps_car_02_cat', 'ps_car_07_cat', 'ps_car_09_cat', 
                'ps_car_14']
impt.randomForestRegression(train, test, columns_miss, 'ps_car_11', fill_train=False)
impt.randomForestRegression(train, train, columns_miss, 'ps_car_11')

# Impute ps_car_07_cat
columns_miss = ['ps_ind_02_cat', 'ps_ind_04_cat', 'ps_ind_05_cat', 'ps_reg_03', 
                'ps_car_01_cat', 'ps_car_02_cat', 'ps_car_09_cat', 
                'ps_car_11', 'ps_car_14']
impt.randomForestRegression(train, test, columns_miss, 'ps_car_07_cat', fill_train=False)
impt.randomForestRegression(train, train, columns_miss, 'ps_car_07_cat')

# Impute ps_ind_02_cat
columns_miss = ['ps_ind_04_cat', 'ps_ind_05_cat', 'ps_reg_03', 
                'ps_car_01_cat', 'ps_car_02_cat', 'ps_car_07_cat', 'ps_car_09_cat', 
                'ps_car_11', 'ps_car_14']
impt.randomForestClassifier(train, test, columns_miss, 'ps_ind_02_cat', fill_train=False)
impt.randomForestClassifier(train, train, columns_miss, 'ps_ind_02_cat')

# Impute ps_car_01_cat
columns_miss = ['ps_ind_02_cat', 'ps_ind_04_cat', 'ps_ind_05_cat', 'ps_reg_03', 
                'ps_car_02_cat', 'ps_car_07_cat', 'ps_car_09_cat', 
                'ps_car_11', 'ps_car_14']
impt.randomForestClassifier(train, test, columns_miss, 'ps_car_01_cat', fill_train=False)
impt.randomForestClassifier(train, train, columns_miss, 'ps_car_01_cat')

# Impute ps_car_12 in test data (only test data has this missing)
columns_miss = ['ps_ind_02_cat', 'ps_ind_04_cat', 'ps_ind_05_cat', 'ps_reg_03', 
                'ps_car_01_cat', 'ps_car_02_cat', 'ps_car_07_cat', 'ps_car_09_cat', 
                'ps_car_11', 'ps_car_14']
impt.randomForestRegression(train, test, columns_miss, 'ps_car_12', fill_train=False)

# Impute ps_car_14
columns_miss = ['ps_ind_02_cat', 'ps_ind_04_cat', 'ps_ind_05_cat', 'ps_reg_03', 
                'ps_car_01_cat', 'ps_car_02_cat', 'ps_car_07_cat', 'ps_car_09_cat', 
                'ps_car_11']
impt.nearbyAverage(train, test, columns_miss, 'ps_car_14', fill_train=False)
impt.nearbyAverage(train, train, columns_miss, 'ps_car_14')

# Impute ps_reg_03
columns_miss = ['ps_ind_02_cat', 'ps_ind_04_cat', 'ps_ind_05_cat', 
                'ps_car_01_cat', 'ps_car_02_cat', 'ps_car_07_cat', 'ps_car_09_cat', 
                'ps_car_11', 'ps_car_14']
impt.randomForestRegression(train, test, columns_miss, 'ps_reg_03', fill_train=False)
impt.randomForestRegression(train, train, columns_miss, 'ps_reg_03')

# Impute ps_car_09_cat
columns_miss = ['ps_ind_02_cat', 'ps_ind_04_cat', 'ps_ind_05_cat', 'ps_reg_03', 
                'ps_car_01_cat', 'ps_car_02_cat', 'ps_car_07_cat', 
                'ps_car_11', 'ps_car_14']
impt.randomForestClassifier(train, test, columns_miss, 'ps_car_09_cat', fill_train=False)
impt.randomForestClassifier(train, train, columns_miss, 'ps_car_09_cat')

# One hot encoding for categorical columns after imputed
columns = ['ps_car_01_cat', 'ps_car_09_cat', 'ps_ind_02_cat', 'ps_ind_05_cat']
train = pd.get_dummies(train, columns=columns)
test = pd.get_dummies(test, columns=columns)s
```
